```python
result_dir = '/content/drive/MyDrive/3_Research_Related/ADHD_Research_Google_Colab/DATA/RESULTS'
filename = result_dir+'/kfold_'
```

```python
def getValue(model, k_hop=1, lr=0.1):
    if model == 'CNN':
        results = pd.read_csv(filename+f'{model}_ndam_epo_50.csv')
    elif lr != 0:
        results = pd.read_csv(filename+f'{model}_k_{k_hop}_ndam_epo_50_lr_{lr}.csv')
    else:
        results = pd.read_csv(filename+f'{model}_k_{k_hop}_ndam_epo_50.csv')
    test = results["Test"]
    train = results["Train"]
    valid = results["Valid"]
    return train, valid, test
```

```python
(train11, valid11, test11) = getValue('GCN', 1, 0.0001)
(train12, valid12, test12) = getValue('GCN', 2, 0.0001)
(train13, valid13, test13) = getValue('GCN', 3, 0.0001)
(train14, valid14, test14) = getValue('GCN', 4, 0.0001)
(train15, valid15, test15) = getValue('GCN', 5, 0.0001)
(train16, valid16, test16) = getValue('GCN', 6, 0.0001)

plt.title('GCN Test Accuracy')
plt.plot(test11, label='K=1', linewidth =1)
plt.plot(test12, label='K=2', linewidth =1)
plt.plot(test13, label='K=3', linewidth =1)
plt.plot(test14, label='K=4', linewidth =1)
plt.plot(test15, label='K=5', linewidth =1)
plt.plot(test16, label='K=6', linewidth =1)

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('GCN_Test.png')
plt.show()
```
